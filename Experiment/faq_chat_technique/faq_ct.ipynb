```python
import spacy
#nlp = spacy.load('ja_ginza')
from nltk import word_tokenize

import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
nlu_file = '/home/iftekhar/amiebot/amieCore/amie_core/core/rasa/data/nlu/amie_nlu_en.md'

def intent_chunks(f_name):
    with open(f_name, encoding='utf-8') as f:
        intents = f.read()
    return intents

def generator(nlu_file):
    collector = {}
    for lines in intent_chunks(nlu_file).split('##'):
        if lines:
            intents = lines.split('-')
            entities = extract_entities(tokenize(intents[1]))
            collector[intents[0].strip()] = entities
    return collector

def extract_entities(query):
    entities = []
    for items in query:
        doc = nlp(items)
        for ent in doc.ents:
            if ent.text:
                entities.append(ent.text)
    entity = list(set(entities))
    if entity:
        return entity
    else:
        return query
```

```python
from nltk.corpus import stopwords
st_words = stopwords.words('english')
```

```python
import string
def remove_punc(words):
    table = str.maketrans('', '', string.punctuation)
    return [w.translate(table) for w in words]

def text_processing(items):
    tokens = [w for w in [lemmer.lemmatize(t) for t in word_tokenize(items)] if w not in st_words]
    return remove_punc([word.lower() for word in tokens])
```

```python
collector = {}
for lines in intent_chunks(nlu_file).split('##'):
    if lines:
        intents = lines.split('-')
        saver = []
        [saver.append(' '.join(text_processing(items))) for items in intents[1:]]
        collector[intents[0].strip().replace('intent:','')] = saver
```

```python
data = list(collector.values())
```
