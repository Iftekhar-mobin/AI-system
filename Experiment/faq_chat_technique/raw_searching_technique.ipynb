```python
import unicodedata
from re import search as match_pattern
def _is_japanese(text):
    for ch in text:
        try:
            name = unicodedata.name(ch)
            if "HIRAGANA" in name \
                    or "KATAKANA" in name:
                return True
        except Exception as e:
            print(str(e))
    return False

def _is_english(text):
    if match_pattern(r'[A-Za-z]', text):
        return True
    else:
        return False

def _is_chinese(text):
    # chinese
    if match_pattern("[\u4E00-\u9FAF]", text):
        return True
    else:
        return False
```
