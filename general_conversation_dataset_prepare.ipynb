```python
import json 
  
f = open('/home/ifte/Downloads/Topical-Chat-master/conversations/train.json') 
data = json.load(f) 
```

```python
[x for x in data.keys()]
```

```python
sentiment = []
collector = []
for i in [x for x in data.keys()]: 
    for items in data[i]['content']:
        collector.append(items['message'])
        sentiment.append(items['sentiment'])
```

```python
# convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-1):
		dataX.append(dataset[i:i+look_back][0])
		dataY.append(dataset[i + look_back])
	return dataX, dataY
```

```python
len(collector)
```

```python
q, a = create_dataset(collector)
```

```python
sentiment[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(zip(q, a, sentiment[1:]), columns=['Que', 'Ans', 'Sent'])
```
