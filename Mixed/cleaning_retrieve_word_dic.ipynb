```python
import _pickle as pickle
import re
```

```python
def load_from_file(file_path):
    with open(file_path, "rb") as input_file:
        data = pickle.load(input_file)
    return data
```

```python
words = load_from_file('/home/ifte/resource/model/retrieve_word_dic.pk')
```

```python
collector = []
garbage = []
for x, y in words.items():
    if len(x) < 3 and bool(re.compile('[a-z]').match(x)):
        garbage.append(x)
    elif len(x) < 2:
        w = re.sub(r'[ぁ-んァ-ン]', '', x)
        w = re.sub(r'[a-z]', '', w)
        w = re.sub(r'[0-9]', '', w)
        w = re.sub(r'\W+', '', w)
        if w:
#             print(x)
            collector.append(w)
        else:
            garbage.append(x)
    else:
        collector.append(x)
```
