```python
import re

def clean_text(text):
    text = re.sub(r'\d+', '', text) 
    text = re.sub(r'\n', '', text)
    text = re.sub(r'「」', '', text)
    text = re.sub(r'「 」', '', text)
    text = re.sub(r'・', '', text)
    return text

cleaned = []

with open('/home/iftekhar/AI-system/Language Detect Model/lang_detect_data/ja_splitted.txt', 'r') as f: 
    for line in f:
        if not line.isspace():
            #if len(line) > 100:
            #    print("Yes")
            #else: 
            #    print("No")
            cleaned.append(clean_text(line))
```
