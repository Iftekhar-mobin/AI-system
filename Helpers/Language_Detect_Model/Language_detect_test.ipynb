```python
def _is_english(text):
    if re.search(r'[A-Za-z]', text):
        return True
```

```python
def _is_japanese(text):
    for ch in text:
        try:
            name = unicodedata.name(ch)
            if "HIRAGANA" in name \
                    or "KATAKANA" in name:
                return True
        except Exception as e:
            print(str(e))
    return Falseprint
```

```python
def getChinese(context):
    context = context.decode("utf-8") # convert context from str to unicode
    filtrate = re.compile(u'[^\u4E00-\u9FA5]') # non-Chinese ulen(word_dic.keys())nicode range
    context = filtrate.sub(r'', context) # remove all non-Chinese characters
    context = context.encode("utf-8") # convert unicode back to str
    return context
```
