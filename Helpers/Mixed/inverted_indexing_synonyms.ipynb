```python
def load_from_file(file_path):
    with open(file_path, "rb") as input_file:
        data = pickle.load(input_file)
    return data
```

```python
index = 0
collector = ''
for parts in query_parts:
    if parts not in vocab_trained.keys() and parts in values_dict.keys():
        try:
            alter = [key for key, val in updated_dict.items() if 
                     [i for i in val if i == parts]]
            collector += alter[0] + ' '
        except KeyError:
            pass
    else:
        collector += parts + ' '
    index += 1
```
